```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(color_codes = True)
```

```python
df = pd.read_csv(r"C:\Users\nakha\OneDrive\Desktop\data science\Automobile.csv")
df.head()
```

```python
# total number of rows and columns
df.shape
```

```python
# drop irrelavant columns
df = df.drop(['weight' , 'origin'],axis =1 )
df.columns
```

```python
# checking for duplicate rows
dupl_rows = df[df.duplicated()]
print("no.of duplicate rows are: " , dupl_rows.shape)
```

```python
#Insight = no duplicate row 

# check for any null values
print(df.isnull().sum())
```

```python
# Insight - 6 null values in hp column

# drop null values
print(df.count())
df= df.dropna()
df.count()
```

```python
# boxplot for cylinders  to horse power
sns.boxplot(x="cylinders" , y= "horsepower" , data = df)
plt.show()
```

```python
#Insight - Outliers at cylinder = 6 and 8 with abnormal horsepower
# with inc in no. of cylinders-> hp increases

#histogram for attributes
plt.hist(df['mpg'],color = 'red' , edgecolor ='black')
plt.show()
```

```python
# Analytical information about mpg attribute 
df['mpg'].describe()
```

```python
# Analytical information about all attribute 
df.describe(include ='all')
```

```python
# finding the outliers based on individual attributes
sns.boxplot(x=df['mpg'])
plt.show()
```

```python
# histogram for no.of cars per brand 
df.name.value_counts().nlargest(40).plot(kind = 'bar' , figsize =(10,5))
plt.show()
```

```python
plt.figure(figsize =(10,5))

#selects only numeric data
corr_btw_attr = df.select_dtypes(include ='number')

#finds the correlation btw attributes
corr_btw_attr = corr_btw_attr.corr()

#heatmaps to display  correlation
# darker the color stronger the correlation
sns.heatmap(corr_btw_attr , cmap=("BrBG") , annot = True)
plt.show()
```

```python
#Insight - 1. cylinders to hp  have a positive corr(0.84) -> as no.of cylinders inc hp inc with a ratio of 0.84
          # 2. hp to mpg  have a negative(-0.78) -> hp inc mpg dec with a ratio of 0.78
```

```python
## Conclusion - 
1. Cylinders(as they affect mpg) and Horsepower clearly shows a trade-Off performance.
2. Cars on the basis of above two can be categorized as economy cars(low hp, high mpg) , family cars(moderate of both), Porche cars(high hp).
3. Companies can  opt for producing moderate specs(optimized mpg , moderate hp) cars to capture the larger segment of market.

## Skills Demonstrated -
1. Exploratory dta analysis skills.
2. Data Visualization
3. Data Wrangling
4. Python Libraries

```
